# About the project

### In this project we will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an advertisement and  to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* Daily Time Spent on Site: consumer time on site in minutes
* Age: customer age in years
* Area income: Avg income of geographical area of consumer
* Daily internet usage: Avg minutes a day consumer is on the internet
* Ad Topic Line: Headline of the advertisement
* City: City of consumer
* Male: Whether or not consumer was male
* Country: Country of consumer
* Timestamp: Time at which consumer clicked on Ad or closed window
* Clicked on Ad: 0 or 1 indicated clicking on Ad

##  Split the Data

```python
df_advert.columns
```

```python
X = df_advert[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male' ]]

y = df_advert['Clicked on Ad']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Training and fit the model

```python
np.random.seed(42)

from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
